### 使用Kmeans聚類分析鳶尾花

KMeans聚類鳶尾花不需要類標（屬於某一類鳶尾花），其是根據資料之間的相似性，按照“物以類聚，人以群分”進行聚類。

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()    
clf = KMeans()  
clf.fit(iris.data, iris.target)  
```

```python
predicted = clf.predict(iris.data)
```

```python
#獲取花卉兩列資料集  
X = iris.data  
L1 = [x[0] for x in X]  
L2 = [x[1] for x in X]
```

```python
plt.scatter(L1, L2, c = predicted, s = 50, cmap='Paired')  
```
