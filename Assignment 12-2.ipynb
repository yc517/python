### 畫出3D圖

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)#生成一個3d物件

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)#對X,Y陣列進行擴充

R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.set_xlabel('X label',color='blue') #設定x座標
ax.set_ylabel('Y label',color='blue')
ax.set_zlabel('Z label')
ax.plot_surface(X, Y, Z,rstride=1,cstride=1,alpha = .6, cmap='rainbow')#生成一個曲面

fig=plt.figure()
```
