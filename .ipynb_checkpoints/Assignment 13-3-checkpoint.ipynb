```python
import tkinter as tk
import __future__
```

```python
window = tk.Tk()
window.title("Calculator")
window.geometry('425x625')

var = tk.StringVar()

l = tk.Label(window, text = '', bg = 'white', font = ('Times', 30), 
            width = 15, height = 3)
l.place(x = 0, y = 0)
txt = ''

def num(x):
    global txt
    txt = txt + x
    l.config(text = txt)

def equal():
    global txt
    #eval(str)：將字符串str當成有效的表達式來求值並返回計算結果
    a = eval(compile(txt, '<string>', 'eval', __future__.division.compiler_flag))
    l.config(text = a)

def clean():
    global txt
    l.config(text = '')
    txt = ''

#符號設置
btnDot = tk.Button(window, text='.', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('.'))
btn1 = tk.Button(window, text='1', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('1'))
btn2 = tk.Button(window, text='2', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('2'))
btn3 = tk.Button(window, text='3', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('3'))
btn4 = tk.Button(window, text='4', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('4'))
btn5 = tk.Button(window, text='5', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('5'))
btn6 = tk.Button(window, text='6', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('6'))
btn7 = tk.Button(window, text='7', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('7'))
btn8 = tk.Button(window, text='8', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('8'))
btn9 = tk.Button(window, text='9', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('9'))
btn0 = tk.Button(window, text='0', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('0'))
btnPlus = tk.Button(window, text='+', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('+'))
btnMinus = tk.Button(window, text='-', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('-'))
btnMultiply = tk.Button(window, text='x', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('*'))
btnDivide = tk.Button(window, text='÷', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = lambda:num('/'))
btnCal = tk.Button(window, text='=', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = equal)
btnClean = tk.Button(window, text='C', bg = 'white', font = ('Times', 15), 
            width = 9, height = 5, command = clean)

#排版place
btn7.grid(row = 1, column = 0)
btn8.grid(row = 1, column = 1)
btn9.grid(row = 1, column = 2)
btnPlus.grid(row = 1, column = 3)
btn4.grid(row = 2, column = 0)
btn5.grid(row = 2, column = 1)
btn6.grid(row = 2, column = 2)
btnMinus.grid(row = 2, column = 3)
btn1.grid(row = 3, column = 0)
btn2.grid(row = 3, column = 1)
btn3.grid(row = 3, column = 2)
btnMultiply.grid(row = 3, column = 3)
btn0.grid(row = 4, column = 0)
btnDot.grid(row = 4, column = 1)
btnCal.grid(row = 4, column = 2)
btnDivide.grid(row = 4, column = 3)
btnClean.grid(row = 0, column = 3)

window.mainloop()
```
